```python
file_2 = 'prob_product_overlapping.npy'
file_1 = 'cumulative_exp_time_sum_overlapping.npy'
file_3 = 'neural_overlapping.npy'
exp_time_overlapping = np.load(input_directory + file_1)
prob_overlapping = np.load(input_directory + file_2)
neural = np.load(input_directory + file_3, allow_pickle=True)
correlation_overlapping = np.zeros((4,20))
correlation_overlapping1 = np.zeros((4,20))
correlation_overlapping2 = np.zeros((4,20))

for i in range(20):
    x = exp_time_overlapping[0][:,i]
    y = prob_overlapping[0][:,i]
    z = neural[0][:,i]
    corr = np.corrcoef(x,y)
    corr1 = np.corrcoef(x,z)
    corr2 = np.corrcoef(y,z)
    if np.isnan(corr[0,1]) == False and np.isnan(corr1[0,1]) == False and np.isnan(corr2[0,1]) == False :
        correlation_overlapping[0,i] = corr[0,1]
        correlation_overlapping1[0,i] = corr1[0,1]
        correlation_overlapping2[0,i] = corr2[0,1]

    x = exp_time_overlapping[1][:,i]
    y = prob_overlapping[1][:,i]
    z = neural[1][:,i]
    corr = np.corrcoef(x,y)
    corr1 = np.corrcoef(x,z)
    corr2 = np.corrcoef(y,z)
    if np.isnan(corr[0,1]) == False and np.isnan(corr1[0,1]) == False and np.isnan(corr2[0,1]) == False :
        correlation_overlapping[1,i] = corr[0,1]
        correlation_overlapping1[1,i] = corr1[0,1]
        correlation_overlapping2[1,i] = corr2[0,1]

    x = exp_time_overlapping[2][:,i]
    y = prob_overlapping[2][:,i]
    z = neural[2][:,i]
    z=np.mean(z.reshape(2,5),axis = 0)
    corr = np.corrcoef(x,y)
    corr1 = np.corrcoef(x,z)
    corr2 = np.corrcoef(y,z)
    if np.isnan(corr[0,1]) == False and np.isnan(corr1[0,1]) == False and np.isnan(corr2[0,1]) == False :
        correlation_overlapping[2,i] = corr[0,1]
        correlation_overlapping1[2,i] = corr1[0,1]
        correlation_overlapping2[2,i] = corr2[0,1]

    x = exp_time_overlapping[3][:,i]
    y = prob_overlapping[3][:,i]
    z = neural[3][:,i]
    z=np.mean(z.reshape(2,5),axis = 0)
    corr = np.corrcoef(x,y)
    corr1 = np.corrcoef(x,z)
    corr2 = np.corrcoef(y,z)
    if np.isnan(corr[0,1]) == False and np.isnan(corr1[0,1]) == False and np.isnan(corr2[0,1]) == False :
        correlation_overlapping[3,i] = corr[0,1]
        correlation_overlapping1[3,i] = corr1[0,1]
        correlation_overlapping2[3,i] = corr2[0,1]
```
